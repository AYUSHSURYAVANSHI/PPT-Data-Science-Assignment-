```python
# 5. Scenario: A study is conducted to compare the performance of two different teaching methods. The test scores of the students in each group are as follows:
#    Group A: [85, 90, 92, 88, 91]
#    Group B: [82, 88, 90, 86, 87]
#    Perform a t-test to determine if there is a significant difference in the mean scores between the two groups.
# Answer :-
import scipy.stats as stats

group_a = [85, 90, 92, 88, 91]
group_b = [82, 88, 90, 86, 87]

# Perform independent t-test
t_statistic, p_value = stats.ttest_ind(group_a, group_b)

# Output the results
print("t-statistic:", t_statistic)
print("p-value:", p_value)
```

```python
# 6. Scenario: A company wants to analyze the relationship between advertising expenditure and sales. The data collected is as follows:
#    Advertising Expenditure (in thousands): [10, 15, 12, 8, 14]
#    Sales (in thousands): [25, 30, 28, 20, 26]
#    Calculate the correlation coefficient between advertising expenditure and sales
# Answer :-
# Step 1: Calculate the mean of the advertising expenditure (X) and sales (Y) data sets:

# Mean of X: (10 + 15 + 12 + 8 + 14) / 5 = 11.8
# Mean of Y: (25 + 30 + 28 + 20 + 26) / 5 = 25.8
# Step 2: Calculate the deviations from the mean for both X and Y:

# Deviations from the mean of X: [10 - 11.8, 15 - 11.8, 12 - 11.8, 8 - 11.8, 14 - 11.8] = [-1.8, 3.2, 0.2, -3.8, 2.2]
# Deviations from the mean of Y: [25 - 25.8, 30 - 25.8, 28 - 25.8, 20 - 25.8, 26 - 25.8] = [-0.8, 4.2, 2.2, -5.8, 0.2]
# Step 3: Calculate the product of the deviations from the mean for X and Y:

# Product of deviations: [-1.8 * -0.8, 3.2 * 4.2, 0.2 * 2.2, -3.8 * -5.8, 2.2 * 0.2] = [1.44, 13.44, 0.44, 22.04, 0.44]
# Step 4: Calculate the sum of the products of the deviations:

# Sum of the product of deviations: 1.44 + 13.44 + 0.44 + 22.04 + 0.44 = 37.8
# Step 5: Calculate the standard deviation of X and Y:

# Standard deviation of X:

# Variance of X = [(10 - 11.8)^2 + (15 - 11.8)^2 + (12 - 11.8)^2 + (8 - 11.8)^2 + (14 - 11.8)^2] / 5 = 5.36
# Standard deviation of X = sqrt(5.36) ≈ 2.314
# Standard deviation of Y:

# Variance of Y = [(-0.8)^2 + (4.2)^2 + (2.2)^2 + (-5.8)^2 + (0.2)^2] / 5 = 15.76
# Standard deviation of Y = sqrt(15.76) ≈ 3.969
# Step 6: Calculate the correlation coefficient (r):

# r = (sum of the product of deviations) / (n * standard deviation of X * standard deviation of Y)
# r = 37.8 / (5 * 2.314 * 3.969) ≈ 0.810
# Therefore, the correlation coefficient between advertising expenditure and sales for the given data is approximately 0.810.
```

```python
# 12. Scenario: A company wants to analyze the click-through rates of two different website designs. The number of clicks for each design is as follows:
#     Design A: [100, 120, 110, 90, 95]
#     Design B: [80, 85, 90, 95, 100]
#     Perform a chi-square test to determine if there is a significant difference in the click-through rates between the two designs.
# Answer :-
import scipy.stats as stats
import numpy as np

design_a = [100, 120, 110, 90, 95]
design_b = [80, 85, 90, 95, 100]

# Prepare the observed frequency table
observed_frequencies = np.array([design_a, design_b])

# Perform chi-square test
chi2_statistic, p_value = stats.chisquare(observed_frequencies, axis=None)

# Output the results
print("Chi-square statistic:", chi2_statistic)
print("p-value:", p_value)
```

```python
# 13. Scenario: A survey is conducted to measure customer satisfaction with a product on a scale of 1 to 10. The data collected is as follows:
#     [7, 9, 6, 8, 10, 7, 8, 9, 7, 8]
#     Calculate the 95% confidence interval for the population mean satisfaction score.
# Answer :-
# Step 1: Calculate the sample mean:

# Sample mean = (7 + 9 + 6 + 8 + 10 + 7 + 8 + 9 + 7 + 8) / 10 = 7.9
# Step 2: Calculate the sample standard deviation:

# Subtract the sample mean from each data point, square the differences, and calculate the average. Finally, take the square root.
# Deviations from the mean: [7 - 7.9, 9 - 7.9, 6 - 7.9, 8 - 7.9, 10 - 7.9, 7 - 7.9, 8 - 7.9, 9 - 7.9, 7 - 7.9, 8 - 7.9]
# Squared deviations: [(-0.9)^2, (1.1)^2, (-1.9)^2, (0.1)^2, (2.1)^2, (-0.9)^2, (0.1)^2, (1.1)^2, (-0.9)^2, (0.1)^2]
# Sample variance = (sum of squared deviations) / (number of data points - 1)
# Sample variance = (0.81 + 1.21 + 3.61 + 0.01 + 4.41 + 0.81 + 0.01 + 1.21 + 0.81 + 0.01) / (10 - 1)
# Sample variance ≈ 2.322 (rounded to three decimal places)
# Sample standard deviation = square root of the sample variance ≈ sqrt(2.322) ≈ 1.523 (rounded to three decimal places)
# Step 3: Determine the sample size (n):

# The sample size is 10 (as given in the scenario).
# Step 4: Calculate the standard error (SE):

# Standard error (SE) = sample standard deviation / square root of the sample size
# SE = 1.523 / sqrt(10) ≈ 0.482 (rounded to three decimal places)
# Step 5: Determine the critical value (z):

# The critical value corresponds to the desired confidence level. For a 95% confidence interval, the critical value is approximately 1.96.
# Step 6: Calculate the margin of error (ME):

# Margin of error (ME) = critical value (z) * standard error (SE)
# ME = 1.96 * 0.482 ≈ 0.946 (rounded to three decimal places)
# Step 7: Calculate the confidence interval:

# Confidence interval = sample mean ± margin of error
# Confidence interval = 7.9 ± 0.946
# Confidence interval ≈ [6.954, 8.846] (rounded to three decimal places)
# Therefore, the 95% confidence interval for the population mean satisfaction score is approximately [6.954, 8.846].
```
